```python
exam.drop(['SEQN'], axis = 1, inplace=True)
demo.drop(['SEQN'], axis = 1, inplace=True)
diet.drop(['SEQN'], axis = 1, inplace=True)
ques.drop(['SEQN'], axis = 1, inplace=True)
```

# Reformatting

Changing Income Variable

```python
def map_income(value):
    if value == 1:
        return "$0-$4999"
    elif value == 2:
        return "$5000-$9999"
    elif value == 3:
        return "$10000-$14999"
    elif value == 4:
        return "$15000-$19999"
    elif value == 5:
        return "$20000-$24999"
    elif value == 6:
        return "$25000-$34999"
    elif value == 7:
        return "$35000-$44999"
    elif value == 8:
        return "$45000-$54999"
    elif value == 9:
        return "$55000-$64999"
    elif value == 10:
        return "$65000-$74999"
    elif value == 12:
        return "$20000 and over"
    elif value == 13:
        return "Under $20000"
    elif value == 14:
        return "$75000-$99999"
    elif value == 15:
        return "$100000 and over"
    else:
        return "Unknown"

# Apply the map_income function to INDHHIN2 and create a new column 'income_range'
df['family income'] = df['INDHHIN2'].apply(map_income)
```

```python
def map_level(value):
    if value < 6 or value == 13:
        return "Less than $25000"
    elif (value >= 6 and value <= 10) or value == 12:
        return "$25000 - $74999"
    elif value == 14 or value == 15:
        return "More than $75000"

df['Income Level'] = df['INDHHIN2'].apply(map_level)
```
